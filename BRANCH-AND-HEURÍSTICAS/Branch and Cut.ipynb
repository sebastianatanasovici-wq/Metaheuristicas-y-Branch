```python
from scipy.optimize import linprog

def branch_and_cut(
    c,
    A,
    b,
    generar_cut,
    es_entera
):

    while True:

        resultado = linprog(
            c,
            A_ub=A,
            b_ub=b,
            method="highs"
        )

        if not resultado.success:
            return None

        solucion = resultado.x

        if es_entera(solucion):

            return solucion

        nueva_restriccion, nuevo_b = generar_cut(solucion)

        A.append(nueva_restriccion)
        b.append(nuevo_b)
```
