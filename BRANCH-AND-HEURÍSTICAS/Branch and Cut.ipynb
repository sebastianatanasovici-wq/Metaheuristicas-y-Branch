Branch and Cut

Extiende Branch and Bound agregando restricciones adicionales llamadas cuts o cutting planes.

Estas restricciones eliminan soluciones fraccionales o inválidas sin eliminar soluciones óptimas válidas.

Gracias a esto, el árbol de búsqueda se reduce y el algoritmo encuentra soluciones más rápido.

```python
from scipy.optimize import linprog

def branch_and_cut(
    c,
    A,
    b,
    generar_cut,
    es_entera
):

    while True:

        resultado = linprog(
            c,
            A_ub=A,
            b_ub=b,
            method="highs"
        )

        if not resultado.success:
            return None

        solucion = resultado.x

        if es_entera(solucion):

            return solucion

        nueva_restriccion, nuevo_b = generar_cut(solucion)

        A.append(nueva_restriccion)
        b.append(nuevo_b)
```
