4. Branch and Cut and Price
Idea Principal

Combina:

Branching, Cuts y Generación de columnas

Es una mezcla avanzada de todos los métodos anteriores.

```python
from scipy.optimize import linprog

def branch_and_cut_and_price(
    c,
    A,
    b,
    es_entera,
    generar_cut,
    pricing_problem,
    criterio_termino,
    max_iter=100
):

    for _ in range(max_iter):

        resultado = linprog(
            c,
            A_ub=A,
            b_ub=b,
            method="highs"
        )

        if not resultado.success:
            return None

        solucion = resultado.x

        if not es_entera(solucion):

            nueva_restriccion, nuevo_b = generar_cut(solucion)

            A.append(nueva_restriccion)
            b.append(nuevo_b)

        nueva_columna, costo_reducido = pricing_problem(
            solucion
        )

        if criterio_termino(costo_reducido):

            break

    return solucion
```
