Branch and Bound

Divide un problema grande en subproblemas más pequeños mediante ramas (branching).
Cada rama representa una posible decisión o solución parcial.

El algoritmo calcula bounds para estimar si una rama puede mejorar la mejor solución encontrada. Si una rama no puede producir una mejor solución, se elimina.

Esto permite reducir enormemente la cantidad de soluciones que deben explorarse.

```python
import heapq

class Nodo:

    def __init__(self, estado, nivel, costo, bound):

        self.estado = estado
        self.nivel = nivel
        self.costo = costo
        self.bound = bound

    def __lt__(self, other):

        return self.bound < other.bound


def branch_and_bound(
    estado_inicial,
    generar_hijos,
    funcion_costo,
    funcion_bound,
    es_solucion
):

    mejor_solucion = None
    mejor_costo = float("inf")

    bounds = []

    raiz = Nodo(
        estado_inicial,
        0,
        funcion_costo(estado_inicial),
        funcion_bound(estado_inicial)
    )

    heapq.heappush(bounds, raiz)

    while bounds:

        nodo = heapq.heappop(bounds)

        if nodo.bound >= mejor_costo:
            continue

        if es_solucion(nodo.estado):

            costo = funcion_costo(nodo.estado)

            if costo < mejor_costo:

                mejor_costo = costo
                mejor_solucion = nodo.estado

            continue

        hijos = generar_hijos(nodo.estado)

        for hijo in hijos:

            nuevo_nodo = Nodo(
                hijo,
                nodo.nivel + 1,
                funcion_costo(hijo),
                funcion_bound(hijo)
            )

            if nuevo_nodo.bound < mejor_costo:

                heapq.heappush(bounds, nuevo_nodo)

    return mejor_solucion, mejor_costo
```
