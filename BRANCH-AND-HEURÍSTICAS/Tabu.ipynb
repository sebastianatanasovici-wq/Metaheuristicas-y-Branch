Tabu Search

Utiliza memoria para evitar volver a soluciones recientemente visitadas.

Mantiene una lista tabu que almacena movimientos o soluciones prohibidas temporalmente.

Esto permite explorar nuevas regiones del espacio de búsqueda y evitar ciclos repetitivos.

```python
def tabu_search(
    solucion_inicial,
    generar_vecinos,
    funcion_objetivo,
    max_iter=100,
    tabu_size=10
):

    actual = solucion_inicial
    mejor = actual

    lista_tabu = []

    for _ in range(max_iter):

        vecinos = generar_vecinos(actual)

        vecinos_validos = [
            v for v in vecinos
            if v not in lista_tabu
        ]

        if not vecinos_validos:
            break

        mejor_vecino = min(
            vecinos_validos,
            key=funcion_objetivo
        )

        actual = mejor_vecino

        if (
            funcion_objetivo(actual)
            < funcion_objetivo(mejor)
        ):

            mejor = actual

        lista_tabu.append(actual)

        if len(lista_tabu) > tabu_size:

            lista_tabu.pop(0)

    return mejor, funcion_objetivo(mejor)
```
