Algoritmos Genéticos

Simulan el proceso de evolución biológica.

Trabajan con una población de soluciones que evoluciona mediante:

selección,
crossover,
mutación.

Las mejores soluciones tienen mayor probabilidad de reproducirse y generar nuevas soluciones de mejor calidad.

```python
import random

def algoritmo_genetico(
    poblacion_inicial,
    funcion_fitness,
    seleccion,
    crossover,
    mutacion,
    generaciones=100,
    prob_mutacion=0.1
):

    poblacion = poblacion_inicial[:]

    for _ in range(generaciones):

        nueva_poblacion = []

        while len(nueva_poblacion) < len(poblacion):

            padre1 = seleccion(
                poblacion,
                funcion_fitness
            )

            padre2 = seleccion(
                poblacion,
                funcion_fitness
            )

            hijo = crossover(padre1, padre2)

            if random.random() < prob_mutacion:

                hijo = mutacion(hijo)

            nueva_poblacion.append(hijo)

        poblacion = nueva_poblacion

    mejor = max(
        poblacion,
        key=funcion_fitness
    )

    return mejor, funcion_fitness(mejor)
```
