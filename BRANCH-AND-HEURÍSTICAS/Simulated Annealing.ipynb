```python
import random
import math

def simulated_annealing(
    solucion_inicial,
    generar_vecino,
    funcion_objetivo,
    temperatura_inicial=100,
    temperatura_minima=0.1,
    alpha=0.95,
    max_iter=1000
):

    actual = solucion_inicial
    mejor = actual

    T = temperatura_inicial

    while T > temperatura_minima:

        for _ in range(max_iter):

            vecino = generar_vecino(actual)

            delta = (
                funcion_objetivo(vecino)
                - funcion_objetivo(actual)
            )

            if delta < 0:

                actual = vecino

            else:

                prob = math.exp(-delta / T)

                if random.random() < prob:

                    actual = vecino

            if (
                funcion_objetivo(actual)
                < funcion_objetivo(mejor)
            ):

                mejor = actual

        T *= alpha

    return mejor, funcion_objetivo(mejor)
```
