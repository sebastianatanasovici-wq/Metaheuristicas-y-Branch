4. Ant

   stá inspirado en el comportamiento real de las hormigas cuando buscan comida.

Las hormigas, al desplazarse, dejan una sustancia química llamada:

Feromona

Las demás hormigas detectan esas feromonas y tienden a seguir los caminos donde la concentración es mayor.

```python
import random
import numpy as np

def ant_colony_optimization(
    construir_solucion,
    funcion_objetivo,
    actualizar_feromonas,
    num_hormigas=10,
    iteraciones=100
):

    mejor_solucion = None
    mejor_costo = float("inf")

    feromonas = {}

    for _ in range(iteraciones):

        soluciones = []

        for _ in range(num_hormigas):

            solucion = construir_solucion(
                feromonas
            )

            costo = funcion_objetivo(solucion)

            soluciones.append(
                (solucion, costo)
            )

            if costo < mejor_costo:

                mejor_costo = costo
                mejor_solucion = solucion

        actualizar_feromonas(
            feromonas,
            soluciones
        )

    return mejor_solucion, mejor_costo
```
