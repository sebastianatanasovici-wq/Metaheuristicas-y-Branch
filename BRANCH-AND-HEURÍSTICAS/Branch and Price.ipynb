Branch and Price

Combina Branch and Bound con generación dinámica de columnas.

En lugar de generar todas las variables del problema desde el inicio, el algoritmo crea únicamente las columnas necesarias durante la ejecución.

Esto permite resolver problemas extremadamente grandes con millones de variables potenciales.

```python
def branch_and_price(
    columnas_iniciales,
    resolver_maestro,
    pricing_problem,
    criterio_termino,
    max_iter=100
):

    columnas = columnas_iniciales[:]

    for _ in range(max_iter):

        solucion_maestra = resolver_maestro(columnas)

        nueva_columna, costo_reducido = pricing_problem(
            columnas,
            solucion_maestra
        )

        if criterio_termino(costo_reducido):

            break

        columnas.append(nueva_columna)

    return columnas
```
